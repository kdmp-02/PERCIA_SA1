```python
#F. Write a program using the words in words.txt to create a word guessing game.

import random
import urllib.request


url = "https://raw.githubusercontent.com/AllenDowney/ThinkPython2/master/code/words.txt"
with urllib.request.urlopen(url) as response:
   words = response.read().decode().splitlines()


def randomword():
    word1 = random.choice(words)    
    return word1

def game(word):
    guesses_left = 10
    guessed_letters = []
    while guesses_left > 0:
        
        print("\nYou have", guesses_left, "guesses left.")

    
        display_word = ""
        for letter in word:
            if letter in guessed_letters:
                display_word += letter + " "
            else:
                display_word += "_ "
        print(display_word)

    
        guess = input(" Guess a letter: ").lower()


        if len(guess) != 1 or not guess.isalpha():
            print("\nPlease enter a single letter.")
            continue
        if guess in guessed_letters:
            print("\nYou already guessed that letter.")
            continue


        guessed_letters.append(guess)

    
        if guess in word:
            print("\nGood guess!")

            if all(letter in guessed_letters for letter in word):
                print("\nCongratulations, you guessed the word", word + "!")
                break
        else:
            print("\nSorry, that letter is not in the word.")
            guesses_left -= 1


    if guesses_left == 0:
        print("\nYou ran out of guesses. The word was", word + ".")
        
word = randomword()
print("I'm thinking of a random word. It has", len(word), "letters.\n")
game(word)
```
